### Про признаки

**Признак** (feature) - это некоторая характеристика объекта/образца/сэмпла.

**Целевой признак** (label) - это тот признак, значение которого нужно спрогнозировать.

Признаки бывают разных видов: *бинарный*, *числовой*, *категориальный*.

С бинарным и числовым признаками всё просто - это просто значение "Да/Нет" и
некоторое число (int или float) соответственно. Категориальный же признак
принимает значение из некоторого множества значений этого признака. Например,
некоторая панда согласно признаку "вид" может принимать одно из трёх значений:
большая, малая и кунг-фу. Согласно же признаку "пол" принимает значение из
множества {самец, самка}.

Целевым признаком может быть также не одно конкретное значение, а даже вектор
значений. Эти значения также могут быть разных видов. Например, мы можем
классифицировать какой-то объект сразу на несколько разных не связанных между
собой классов. В таком случае это будет многомерная классификация.

### Какие бывают части набора данных

Чтобы заставить модель машинного обучения / нейросеть решать задачу, нужно до
этого обучить её на некотором куске данных, чтобы потом она хорошо себя
показывала на всех остальных данных.

Все данные (это ещё называют **выборкой**, **датасетом**) делятся на **тренировочные (train)**
и **тестовые (test)**. Пропорции деления можно выбрать какие угодно и
поэкспериментировать - в зависимости от них результаты могут отличаться. Но
наиболее часто делят в пропорциях 70% train и 30% test.

Отличие этих данных заключается в том, что в рамках задачи обучения с учителем
тренировочная выборка содержит в себе колонку целевого признака - того, который
мы будем предсказывать. А тестовая выборка уже эту колонку не содержит, и
подразумевается, что на этой выборке будет осуществляться тестирование модели.

Итак, всё логично - на тренировочных данных модель тренируется, на тестовых -
тестируется.

## Чем работает Pandas

```python
import pandas as pd
import numpy as np
```

Основными структурами данных в Pandas являются классы Series и DataFrame. Первый
из них представляет собой одномерный индексированный массив данных некоторого
фиксированного типа. Второй – это двухмерная структура данных, представляющая

представлять её как словарь объектов типа Series. Структура DataFrame отлично

описаниям отдельных объектов (тех самых сэмплов), а столбцы соответствуют признакам.

Теперь давай посмотрим, где в процентном соотношении больше всего отсутствует
данных. Напишем для этого простенькую функцию, которая будет нам определять это
процентное соотношение.

```python
def check_missing_data(all_data):
  missing_data = {}
  for key in all_data.columns:
      if all_data[key].isnull().sum() > 0:
          missing_data[key] = (all_data[key].isnull().sum()  /  len(all_data[key])) * 100
  return missing_data

missing_data = check_missing_data(all_data)
missing_data = pd.DataFrame(missing_data, index=['MissingValues']).T.sort_values(by='MissingValues',ascending=False)
```

## Простыми словами о важной статистике

Прежде чем перейти к тому, что пригодится сейчас нам на месте, давай в целом
пройдёмся по важным элементам этой науки, которые помогут нам в дальнейшем.

<!-- здесь по хорошему вставить реальные результаты подсчёта, но может и всё равно -->
Пусть у нас есть некоторое яблоко. Я знаю сколько оно весит, а вот 5
опрашиваемых мною людей - нет. Теперь попробуем, не обладая весами, вычислить
сколько оно весит за счёт статистики.

Для этого дадим в руку каждому из 5-х человек это яблоко и попросим угадать его
вес (в граммах). Получим 7 значений:

[176, 139, 100, 200, 120, 120, 140]

Упорядочим их:

[100, 120, 120, 139, 140, 176, 200]

Это - **вариационный ряд**. Он представляет собой значения некоторой переменной
в статистике и обязательно должен быть упорядоченным.

Теперь на примере этого ряда разберем важные определения:

**Среднее значение (mean)** - тут всё просто: всё складываем и делим на количество
элементов. Получаем $142.14$.

**Мода (mode)** - пока тоже всё понятно: самое часто встречающееся значение.
Здесь это $120$.

**Дисперсия (variance)** - это величина, которая позволят понять, насколько
сильно значения этого ряда отличаются от среднего. Чем больше эта величина, тем
сильнее данные отличаются друг от друга. Здесь это значение $1037.83$. Но это
много или мало? Дело в том, что поскольку эта величина относительная, то и
понять, большая она или маленькая, можно понять только относительно другой выборки.

То есть если мы разделим эту выборку на две, мы можем оценить то, насколько
разброс данных в одной из выборок отличается от того, насколько сильно
разбросаны данные в другой выборке:

Дисперсия [176, 139, 100, 200]: 1435

Дисперсия [120, 120, 140]: 88

И сразу становится понятно, что данные в первой выборке гораздо разношерстнее,
чем данные во второй. Но что если нам всё равно на разницу в разбросе с другой
выборкой и нам важно понять, как конкретное значение в нашей выборке может
отличаться от среднего значения? Для этого нам может быть полезно...

**Среднее квадратичное отклонение (std)** - корень из значения дисперсии.

**Медиана (median)** - тоже всё понятно, это срединное значение вариационного
ряда. В нашем случае это $139$. Это означает 50% данных меньше значения $139$, и
50% данных больше.

Но что если нам нужно смотреть не срединное значение, а значение, которое,
допустим, больше чем 25% элементов выборки? Или 75%? Тогда понятие медианы можно
выразить через слово ''квантиль''.

**Квантиль** - это значение, которое разделяет упорядоченный набор данных на
равные части. Другими словами, квантиль определяет, какое значение из набора
оказывается ниже определенной доли данных. Обычно квантиль записают в виде
''$0.x$-квантиль'', где $x$ - это объём выборки в процентах. То есть медиана -
это $0.5$-квантиль.

В случае нашего набора данных давайте для закрепления определим 0.25 и 0.75-квантили:

25% - $\frac{120 + 120}{2} = 120$ 

75% - $\frac{140 + 176}{2} = 158$


Ну и напоследок давайте про что-то интересное. Как определить, насколько сильно
две, три или ещё больше переменных связаны (линейно зависимы) между собой? То
есть, например, что мне нужно знать, чтобы понять, насколько сильно изменится
значение $a_1$, если было увеличено значение $a_2$. Оно уменьшится или
увеличится?

Для понимания этой зависимости между переменными существует **корреляция**. Если
две переменные имеют высокую корреляцию, то изменения в одной переменной обычно
сопровождаются подобными изменениями в другой переменной. К примеру, если есть
положительная корреляция между количеством часов, проведенных учениками за
уроками, и их успеваемостью, то ученики, проводящие больше времени учебе,
склонны иметь более высокие оценки.

#### One-hot encoding  (функция get_dummies)

One-hot encoding (иногда называется "one-of-K encoding") - это метод
представления категориальных данных в виде бинарных векторов.

Функция кодирует все виды признаков в булевые. Важно, чтобы эти признаки были категориальными

```python
all_df_copy = pd.get_dummies(all_data.drop(['Fare','Pclass','Age','Parch','SibSp','Lastname'],axis=1), drop_first=True)
all_df_copy
```

Чем такое кодирование полезно:

* Многие алгоритмы машинного обучения требуют, чтобы данные были числовыми.
  Категориальные переменные, такие как тип жилья (например, апартаменты или
  дом), могут быть преобразованы в форму, понятную компьютеру.

* Многие алгоритмы машинного обучения, такие как линейная регрессия или деревья
  решений, могут работать лучше, когда категориальные переменные представлены в
  виде дамми переменных.

* Преобразование категориальных переменных в дамми переменные может сделать
  результаты модели более интерпретируемыми, так как позволяет явно увидеть
  влияние каждой категории.

Однако стоит помнить, что one-hot encoding может увеличить размерность данных,
особенно если у вас есть много категорий.

# Matplotlib

Matplotlib — это библиотека двумерных графиков, которая помогает визуализировать фигуры. Matplotlib эмулирует графики и визуализации из Matlab (если кто-то вообще слышал об этой штуке). Matlab не бесплатен, его трудно масштабировать, и как язык программирования он утомителен.
```python
import matplotlib.pyplot as plt
```

### Основные типы графиков

* plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
* hist — гистограмма, показывающая распределение некоторой величины;
* scatter — график, показывающий взаимосвязь двух величин;
* bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

#### Plot (lineplot, поточечный линейный график)

```python
# Рассмотрим на примере некоторой функции kx + b
x = [x for x in range(-1000, 1000)]
y = [5 * y + 10 for y in range(-1000, 1000)]

plt.plot(x, y, label='kx + b')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Линейный график довольно часто применяется при анализе временных рядов. Временной ряд это собранный в разные моменты
времени статистический материал о значении каких-либо параметров (в простейшем случае одного) исследуемого процесса.
Временные ряды (Time series) можно представить линейным графиком, который в основном соединяет точки данных прямой
линией. Это полезно для понимания тенденции с течением времени. Это может объяснить корреляцию (то есть зависимость)
между точками по тренду. Тренд - это основное направление изменения какого-то параметра. Восходящий тренд означает
положительную корреляцию, а нисходящий тренд означает отрицательную корреляцию. Он в основном используется в
прогнозировании (предугадывании следующего значения временного ряда), моделях мониторинга (здесь подразумевается анализ качества чего-либо в зависимости от принятых решений по отношению к объекту, качество которого анализируется).

Когда использовать: временные ряды следует отображать с помощью линейного графика, когда необходимо отобразить одну или несколько переменных во
времени.

Например: анализ фондового рынка компаний, прогноз погоды.

#### Hist (Гистограмма)

Гистограмма берет ряд (series) данных и делит данные на несколько ячеек (bins). Затем он отображает частоту
точек данных в каждом бине (в виде интервала точек). Это полезно для понимания количества значений конкретных диапазонов данных.

Когда использовать: мы должны использовать гистограмму, когда нам нужно посчитать количество значений переменных-признаков на графике.

Например: количество людей определенных возрастов.

```python
# Построим распределение людей по возрасту
plt.figure(figsize=(12, 6))
plt.hist(df['Age'], label="Age")
plt.legend()
plt.show()
```

#### Pie (Круговая диаграмма)

Это круговой график, разделенный на срезы, чтобы проиллюстрировать числовую пропорцию. Срез круговой диаграммы
показывает долю частей относительно общего количества.

Когда использовать: круговая диаграмма редко используется, так как бывает трудно сравнивать части диаграммы (особенно
когда количество возможных и встречающихся значений признака большое). Вместо этого используется гистограмма, там
сравнивать разделы полегче

```python
data = [20.0, 10.0, 5.0, 1.0, 0.5]

# Нарисовать круговой график
plt.pie(data)
plt.show()
```

#### Scatter (Точечная диаграмма)

Точечная диаграмма помогает визуализировать 2 числовые переменные (по осям x и y). Это помогает определить взаимосвязь
данных с каждой переменной, т. е. закономерности корреляции или тренда. Обычно полезно для обнаружения выбросов на
графике.

Когда использовать: в концепциях машинного обучения, таких как регрессия, где x и y — непрерывные
переменные. Также полезен для анализа кластеризации разбросов или обнаружения выбросов.

```python
plt.scatter(df['Fare'], df['Age'])
plt.show()
```

#### Bar (...колоночный график?..)

Bar plot показывает распределение данных по нескольким группам. Обычно его путают с гистограммой, которая использует
только числовые данные для построения графика. Это помогает в сравнении нескольких числовых значений.

Когда использовать: используется при сравнении нескольких групп.

Например: порты, на которых пассажиры загружались на титаник.

```python
# Распределение мужчин и женщин по возрасту
plt.bar(df['Embarked'], df['Age'])
plt.show()
```

#### Boxplot (ящик с усами)

Boxplot дает хорошую сводку об анализируемых данных. Это помогает лучше понять распределение выборки.

Когда использовать: его применяют, когда нам требуется применять общую статистическую информацию о
распределении данных. Также полезен для обнаружения выбросов.

<img src="pic/boxplot.png" alt="Drawing" style="width: 250px;"/>


Пояснение ящика с усами

1. Медиана и другие квартили.

На приведенной выше диаграмме линия, которая делит поле на 2 части, представляет собой
медиану данных. Конец прямоугольника показывает верхний квартиль (75%), а начало прямоугольника представляет нижний
квартиль (25%). Верхний квартиль также называют 3-м квартилем, а нижний квартиль также называют 1-м квартилем. Область
между нижним квартилем и верхним квартилем называется межквартильным диапазоном (IQR) и используется для аппроксимации
50-процентного разброса средних данных (75–25 = 50 %).

2. Максимум и минимум.

Максимум — это самое высокое значение в данных, и аналогично минимальное — это самое низкое значение в данных, его также
называют шапкой.

3. Выбросы.

Точки вне прямоугольников и между максимумом и максимумом называются усами, они показывают диапазон значений в данных.
Экстремальные точки являются выбросами данных. Обычно используемое правило заключается в том, что значение является
выбросом, если оно меньше нижнего квартиля - 1,5 * IQR или выше верхнего квартиля + 1,5 * IQR.

Пример: значения возрастов пассажиров разных полов. Мы можем получить максимум, минимум и много другой информации об этих признаках.

```python
sns.boxplot(df['Age'])
sns.boxplot('Sex', 'Age', data=df)
```

#### sns.countplot (по сути Bar plot)

#### Heatmap (тепловая карта)

Тепловая карта (также известная как heatmap) отображает факт зависимости интересующих параметров выборки по двум переменным оси в виде сетки цветных квадратов. Признаки разделены на диапазоны каждой оси, и цвет каждой ячейки указывает значение корреляции (или степень зависимости) признака на оси "x" с признаком на оси "y".

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap='Greens')
```

#### Violinplot (скрипичный график)

Скрипичный график является более лучшей диаграммой, нежели ящик с усами, поскольку он дает гораздо более широкое

которое с помощью boxplot'а не видно.

Когда использовать: это аналог box plot, значит и применяется для тех же целей. Ещё скрипичный график следует
использовать, когда нам требуется лучшее интуитивное понимание данных.

# Оформление графиков

Базовые принципы построения приличных графиков:
- оси должны быть подписаны, причём не слишком мелко;
- у графика должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах);
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать;
- если распределение на гистограмме имеет тяжёлые хвосты, лучше использовать логарифмическую шкалу.

Например:

```python
# вертим размеры графика, как мы хотим
fig_dims = (12, 6)
fig, ax = plt.subplots(figsize=fig_dims)

sns.set_theme(style="whitegrid")
# заходит как-то в bar улитка и говорит бармену...
ax = sns.barplot(y='Age', x='Sex', ax=ax, data=df)

# масштабируем всякие прелести (если очень хочется)
ax.set_ylim(df['Age'].min(), df['Age'].max())

# подписываем оси и сам график
ax.set_title('Столбцы для пола и возраста', fontsize=20)
ax.set_xlabel('Пол', fontsize=20)
ax.set_ylabel('Возраст', fontsize=20)

# при необходимости можно сохранять всё это дело как тебе угодно (в каком угодно формате)
# plt.savefig('plot.pdf')
```
